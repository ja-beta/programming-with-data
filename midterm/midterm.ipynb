now let's try to do the same thing for all categories

```python
job_cats = ['Engineering, Architecture, & Planning', 'Health', 'Legal Affairs',
       'Technology, Data & Innovation', 'Administration & Human Resources',
       'Social Services', 'Finance, Accounting, & Procurement',
       'Engineering, Architecture, & Planning Public Safety, Inspections, & Enforcement',
       'Building Operations & Maintenance',
       'Public Safety, Inspections, & Enforcement',
       'Constituent Services & Community Programs',
       'Constituent Services & Community Programs Health Policy, Research & Analysis',
       'Policy, Research & Analysis',
       'Administration & Human Resources Social Services',
       'Health Public Safety, Inspections, & Enforcement',
       'Health Policy, Research & Analysis',
       'Finance, Accounting, & Procurement Policy, Research & Analysis',
       'Legal Affairs Public Safety, Inspections, & Enforcement',
       'Communications & Intergovernmental Affairs',
       'Engineering, Architecture, & Planning Policy, Research & Analysis',
       'Constituent Services & Community Programs Health',
       'Technology, Data & Innovation Social Services',
       'Constituent Services & Community Programs Social Services',
       'Health Technology, Data & Innovation']
```

```python
for category in job_cats:
    cat_df = df[df["Job Category"] == category]
    
    cat_salaries = pd.concat([
    cat_df["Salary Range From"],
    cat_df["Salary Range To"]
    ])
    
    cat_salaries.plot(kind="box")

    plt.title(f'Salary distribution for {category}')
    plt.ylabel("Salary")

    plt.show()
```

Now one last try to plot all categories together!
First, create an empty dataframe

```python
salary_data = pd.DataFrame()
```

```python
for category in job_cats:
    cat_df = df[df["Job Category"] == category]
    cat_salaries = pd.concat([
    cat_df["Salary Range From"],
    cat_df["Salary Range To"]
    ])
    salary_data[category] = cat_salaries.reset_index(drop=True)
```

```python
salary_data
```

```python
flierprops = dict(marker='.', markersize=2, linestyle='none', markeredgecolor='k')

plt.figure(figsize=(16, 6))
salary_data.boxplot(flierprops=flierprops)

plt.title("Salary Distribution by Job Category")
plt.ylabel("Salary")
plt.xlabel("Job Category")
plt.xticks(rotation=90)

plt.show()
```
