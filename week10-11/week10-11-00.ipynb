# The HAVING clause
It works like the WHERE clause except it works on the results of the data.
WHERE filters before GROUP BY

````python
%%sql
SELECT * FROM river LIMIT 10
````

````python
%%sql
SELECT sea, count(name) FROM river GROUP BY sea
````

````python
%%sql
SELECT sea, count(name) FROM river WHERE source_latitude > 60 GROUP BY sea
````

to do the filtering after the grouping we need the HAVING clause. It takes the results of the GROUP BY and filters that, rather than the rows themselves like WHERE does.
table -> where -> table(filtered rows) -> group by (aggregation) -> aggregation table -> having(filter that table)

````python
%%sql
SELECT sea, count(name) FROM river GROUP BY sea HAVING count(name) > 10
````

````python
%%sql
SELECT * FROM religion LIMIT 10
````

````python
%%sql
SELECT name, count(country) FROM religion GROUP BY name ORDER BY count(country) DESC
````

````python
%%sql
SELECT * FROM religion WHERE name = "Mormon"
````

````python
%%sql
SELECT name, count(country) FROM religion GROUP BY name HAVING count(country) = 1
````

````python
%%sql
SELECT name, count(country) 
FROM religion 
WHERE percentage < 5
GROUP BY name 
HAVING count(country) = 1
````

# table joins
when you have relational tables, data is usually related using unique keys.
So there's usually an ID field.
So even if all the information is the same, each entry still has a unique identifier.
Using the ID, we can also reference that entry in a different table.
In our current dataset the keys are the country codes, although usually they're just sequential numbers or random numbers

* "unique keys"
    * primary key

Joining only works if we have a unique key. Otherwise it'll take the first one it finds that matches the criteria.

````python
%%sql
SELECT name, country FROM city
````

````python
%%sql
SELECT * FROM country
````

````python
%%sql
SELECT name FROM country WHERE code = "NL"
````

left_table, right_table

    SELECT fields
    FROM left_table
    JOIN right_table ON left_field = right_field

This will compare the fields from both columns, pull the rows from the left table that match the stuff in the right table and join them.
In the example below the "code" is not part of the syntax, it's the name of the table we're looking at.

````python
%%sql
SELECT city.name, city.population, country.code, country.name, country.population
FROM city
JOIN country ON city.country = country.code
````

````python
%%sql
SELECT * FROM airport LIMIT 5
````

````python
%%sql
SELECT airport.IATACode, airport.name, country.name
FROM airport JOIN country ON airport.country = country.code
ORDER BY airport.IATACode
````

````python
%%sql
SELECT airport.IATACode, airport.name, country.name, country.population
FROM airport JOIN country ON airport.country = country.code
ORDER BY airport.IATACode
````

````python
%%sql
SELECT airport.IATACode, airport.name, country.name, country.population, airport.elevation
FROM airport JOIN country ON airport.country = country.code
WHERE country.population < 5000000 AND airport.elevation > 1000
ORDER BY airport.IATACode
````

````python
%%sql
SELECT country.name, sum(city.population)
FROM city JOIN country ON city.country = country.code
GROUP BY country.name
````

````python
%%sql
SELECT country.name, sum(city.population)
FROM city JOIN country ON city.country = country.code
GROUP BY city.country
HAVING sum(city.population) > 20000000
ORDER BY sum(city.population) DESC
````

A country can have multiple cities, whereas citys can only be in one country. So this is a many to one relationship.

# Many to many
for example, the geo_river is a linking table that joins rivers with countries!

````python
%%sql
SELECT * FROM geo_river
````

````python
%%sql
SELECT geo_river.river, country, country.name, geo_river.province 
FROM geo_river JOIN country ON geo_river.country = country.code
WHERE river = 'Rhein'
````

````python
%%sql
SELECT * FROM river
````

````python
%%sql
SELECT river.name, river.length, country.name, geo_river.province
FROM river 
JOIN geo_river ON river.name = geo_river.river
JOIN country ON country.code = geo_river.country
WHERE river.length > 4000
````
