Histograms with floating number (rather than integers) is different:
it looks at ranges and divides the data into bins - equally spaced areas along the x axis,

# Grouping!
We would usually want to use groupby by values that are discrete and not continuous (like hours, months...)

```python
df
```

```python
df[df["month"] == 1] #a dataframe that only has data where month is 1
```

```python
df[df["month"] == 1]["pm2.5"].mean() #average of pm across january
```

```python
df[df["month"] == 2]["pm2.5"].mean() #same for february
```

```python
df.groupby("month")
```

```python
#evaluates to this dataframe groupby object... it lets us take the values and perform aggregation on it (mean, max, std...)
```

```python
df.groupby("month").mean(numeric_only=True) #get average for every column by month
```

```python
df.groupby("month").mean(numeric_only=True)["Iws"] #just for a particular column of the aggregation
```

```python
df.groupby("hour").max(numeric_only=True) #the max for each field by day
```

```python
df.groupby("hour").max(numeric_only=True)["pm2.5"] #just for polution - max polution by hour of day
```

```python
df.groupby("hour").max(numeric_only=True)["pm2.5"].plot()
```

```python
mean_by_month_df = df.groupby("month").mean(numeric_only=True)
```

```python
mean_by_month_df
```

```python
mean_by_month_df[mean_by_month_df["Is"] > 0] #the months where the average snowfall was greater than 0
```

```python
mean_by_month_df.sort_values(by="pm2.5", ascending=False) #order the months by average pollution value
```

```python
mean_by_month_df.plot(kind="bar", y=["pm2.5", "Iws", "TEMP"])
```

## column data types

```python
museum_data = """\
Number,Street,Zip,Price
200,Central Park West,10024,$28
200,Eastern Parkway,11238,$20
1071,Fifth Avenue,10128,$30
1000,Fifth Avenue,10028,$25
11,West 53rd Street,10019,$25"""

from io import StringIO
museum_df = pd.read_csv(StringIO(museum_data))
```

```python
museum_df
```

```python
museum_df.info()
```

```python
# two bad guesses - zipcodes are not integers (0 is important to have), and price should be number.
# so we want to be able to fix that
# available types are: int, bool, float, str
```

```python
museum_df["Zip"].astype(str)
```

```python
museum_df.assign(zip_str=museum_df["Zip"].astype(str))
```

```python
# For the price we can't just change the type, we have to remove the "$"
```

```python
fixed_museum_df = museum_df.assign(price_int=museum_df["Price"].str[1:].astype(int)) 
#give me everything after the first character, convert to int, add a column and save as a new dataframe
```

```python
fixed_museum_df
```

```python
fixed_museum_df["price_int"].mean()
```
