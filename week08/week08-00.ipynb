# correlation study

```python
df = pd.DataFrame(
    {
        "a": np.linspace(0, 100, 50),
        "b": np.linspace(0, 60, 50) + np.sin(np.linspace(0, 40, 50)) * 10,
        "c": np.linspace(200, 75, 50) + np.random.uniform(-5, 5, 50),
        "d": np.random.uniform(0, 25, size=50)
    }
)
```

```python
df.plot()
```

```python
df.plot(kind="scatter", x="a", y="b")
```

```python
df.plot(kind="scatter", x="a", y="d")
```

```python
# randomizing the order of the array. Because order is not neccessarily meaningful!
df = df.sample(len(df)).reset_index(drop=True)
```

```python
df.plot()
```

```python
# but in a scatter plot it would look the same
df.plot(kind="scatter", x="a", y="b")
```

```python
df.plot(kind="scatter", x="c", y="a")
```

```python
df.plot(kind="scatter", x="c", y="b")
```

```python
df.plot(kind="scatter", x="c", y="d") #anything with d will give no correlation
```

```python
# a handy tool for investigating correlation:
pd.plotting.scatter_matrix(df)
# instead of just showing a line in the diagonal axis its giving a histogram of the data
```

## actual correlation values!

```python
df.corr() #the pearson correlation coefficient - prints out a correlation matrix - a numerical score from -1 to 1 - negative to positive correlation
```

# clustering!
## dimensional reduction!!
techniques that try to preserve the correlation between rows and columns while reducing the number of columns in the dataset.

the algorithm we use is called: t-SNE 

```python
df_topics = df.T
df_econ = df
```

```python
# import library for t-SNE
import sys
!{sys.executable} -m pip install scikit-learn
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, n_iter=3000).fit_transform(df_topics) #reduce all data to 2 columns
```

```python
tsne[:10]
```

```python
tsne_df = pd.DataFrame(tsne, index=df_topics.index, columns=["x", "y"])
```

```python
tsne_df.plot(kind="scatter", x="x", y="y")
```

```python
fig, ax = plt.subplots() # create a matplotlib axis
tsne_df.plot.scatter(0, 1, ax=ax, figsize=(8, 8)) # figsize sets the size of the figure
for k, v in tsne_df.iterrows(): # annotate with matplotlib for every row
    ax.annotate(k, v)
```
